#### Manually calculate standard deviation

$$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

- Write a function to calculate standard deviation using the formula above

````python
#code

def standard_deviation_function (attribute_name):
    n = len(combined_2017_2018[[attribute_name]])
    
    total = 0
    variance = 0 
    
    for i in range(n):
        total = total + combined_2017_2018.loc[i, [attribute_name]]
        
    mean = total / n
    
    for j in range(n):
        variance += ((combined_2017_2018.loc[j, [attribute_name]] - mean) ** 2)/(n)

      
    return variance ** .5
````

- Use a **dictionary comprehension** to apply your standard deviation function to each numeric column in the dataframe.  **No loops**  
- Assign the output to variable `sd` as a dictionary where: 
    - Each column name is now a key 
    - That standard deviation of the column is the value 
     
*Example Output :* `{'ACT_Math': 120, 'ACT_Reading': 120, ...}`

````python
#Code:

std_combined_2017_2018 = { 
    'act_participation' : standard_deviation_function("act_participation"),
    'act_english' : standard_deviation_function("act_english"),
    'act_math' : standard_deviation_function("act_math"),
    'act_reading' : standard_deviation_function("act_reading"),
    'act_science' : standard_deviation_function("act_science"),
    'act_composite' : standard_deviation_function("act_composite"),
    'sat_read_write' : standard_deviation_function("sat_read_write"),
    'sat_math' : standard_deviation_function("sat_math"),
    'sat_total' : standard_deviation_function("sat_total"),
    'year' : standard_deviation_function("year")
}

display(std_combined_2017_2018)
````
